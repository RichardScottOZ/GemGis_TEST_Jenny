```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline

```

```python
xls = pd.ExcelFile(os.path.join('cptu.xlsx'))
cpt11 = pd.read_excel(xls, 'cptu11',header = [0])
cpt12 = pd.read_excel(xls, 'cptu12',header = [0])
cpt13 = pd.read_excel(xls, 'cptu13',header = [0])
cpt14 = pd.read_excel(xls, 'cptu14',header = [0])
cpt14A = pd.read_excel(xls, 'cptu14A',header = [0])
cpt15 = pd.read_excel(xls, 'cptu15',header = [0])
cpt16 = pd.read_excel(xls, 'cptu16',header = [0])
cpt17 = pd.read_excel(xls, 'cptu17',header = [0])
cpt20 = pd.read_excel(xls, 'cptu20',header = [0])
```

```python
cpt14A.head()
```

```python
legend = xls.sheet_names
```

```python
legend
```

```python
cpt = [cpt11, cpt12, cpt13, cpt14, cpt14A, cpt15, cpt16, cpt17, cpt20]
```

```python
plt.rc('font', size=15) 
```

```python
for  count, c in enumerate(cpt):
    ocr = c.OCR
    elev = c.elevacao
    fig0 = plt.figure(0,figsize=(5,10))
    plt.plot(ocr,elev,'*',label = legend[count])
    plt.xlabel('OCR')
    plt.ylabel('Elevação[m]')
    plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")

plt.show()
    
    
```

```python
for  count, c in enumerate(cpt):
    qc = c.qc
    elev = c.elevacao
    fig0 = plt.figure(0,figsize=(5,10))
    plt.plot(qc,elev,label = legend[count])
    plt.xlabel('Resistência de Ponta - qc [MPa]')
    plt.ylabel('Elevação[m]')
    plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")

plt.show()
```

```python
for  count, c in enumerate(cpt):
    qt = c.qt
    gamma = c.gamma
    fig0 = plt.figure(0,figsize=(5,10))
    plt.plot(qt,gamma,'*',label = legend[count])
    plt.xlabel('Resistência de Ponta corrigida - qt [MPa]')
    plt.ylabel('$\gamma$ [kN/$m^3$]')
    plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")

plt.show()
```

```python
for  count, c in enumerate(cpt):
    Ic = c.Ic
    elev = c.elevacao
    fig0 = plt.figure(0,figsize=(5,10))
    plt.plot(Ic,elev,'*',label = legend[count])
    plt.xlabel('Ic')
    plt.ylabel('Elevação [m]')
    plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")

plt.show()
```
